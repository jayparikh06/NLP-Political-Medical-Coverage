```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/environ_mm.csv")
```

```python
sampled = df[df['relevance']==1]
```

```python
df_large = pd.read_excel("data/american_2021_environment_compiled_headlines_subsample_large.xlsx")
```

```python
compiled_df = pd.merge(sampled,df_large, left_on=['url'],right_on=['url'])
```

```python
#list of all news sources that were collected from GDELT
source_stance = {'apnews.com':'C',
'breitbart.com':'R',
'cbsnews.com':'SL',
'chicagotribune.com':'C',
'chron.com':'C',
'cnn.com':'L',
'dailykos.com':'L',
'foxnews.com':'R',
'latimes.com':'SL',
'msnbc.com':'L',
'nbcnews.com':'SL',
'newsweek.com':'C',
'newyorker.com':'L',
'npr.org':'SL',
'nydailynews.com':'L',
'nymag.com':'L',
'nytimes.com':'SL',
'politico.com':'SL',
'qz.com':'C',
'reuters.com':'C',
'salon.com':'L',
'sfgate.com':'SL',
'slate.com':'L',
'theatlantic.com':'SL',
'theblaze.com':'R',
'thedailybeast.com':'L',
'time.com':'SL',
'upworthy.com':'L',
'usatoday.com':'SL',
'vox.com':'L',
'washingtonpost.com':'SL',
'zerohedge.com':'SR'}
```

```python
compiled_df['source_stance'] = compiled_df['source'].apply(lambda x: source_stance[x])
```

```python
compiled_df.groupby(by=['source_stance'])[['relevance', 'pro_environment',
       'extremeweatherevent', 'renewableenergy', 'nonrenewableenergy',
       'wildfire', 'pollution', 'ecosystem', 'sealevelrise', 'politics',
       'finance', 'publichealth', 'lifestyle', 'fear', 'hope', 'numbers',
       'gpe', 'person', 'org']].sum()
```

```python
compiled_df.groupby(by=['source_stance'])[['relevance', 'pro_environment',
       'extremeweatherevent', 'renewableenergy', 'nonrenewableenergy',
       'wildfire', 'pollution', 'ecosystem', 'sealevelrise', 'politics',
       'finance', 'publichealth', 'lifestyle', 'fear', 'hope', 'numbers',
       'gpe', 'person', 'org', 'positive', 'negative', 'neutral', 'compound']].mean()
```

```python
source_stance_likert={'L':-2, 'SL':-1, 'C':0, 'SR':1, 'R':2}
compiled_df['source_stance_likert'] = compiled_df['source_stance'].apply(lambda x: source_stance_likert[x])
```

```python
import numpy as np
import statsmodels.api as sm
```

```python
subset = compiled_df[['pro_environment',
       'extremeweatherevent', 'renewableenergy', 'nonrenewableenergy',
       'wildfire', 'pollution', 'ecosystem', 'sealevelrise', 'politics',
       'finance', 'publichealth', 'lifestyle', 'fear', 'hope', 'numbers',
       'gpe', 'person', 'org', 'compound','source_stance_likert']].dropna()


x = subset[['pro_environment',
       'extremeweatherevent', 'renewableenergy', 'nonrenewableenergy',
       'wildfire', 'pollution', 'ecosystem', 'sealevelrise', 'politics',
       'finance', 'publichealth', 'lifestyle', 'fear', 'hope', 'numbers',
       'gpe', 'person', 'org', 'compound']]
y = subset['source_stance_likert']
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
print(results.summary())
```

```python
compiled_df['source_stance_likert_three'] = compiled_df['source_stance_likert'].apply(lambda x: 0 if x>0 else x)
```

```python
subset = compiled_df[['pro_environment',
       'extremeweatherevent', 'renewableenergy', 'nonrenewableenergy',
       'wildfire', 'pollution', 'ecosystem', 'sealevelrise', 'politics',
       'finance', 'publichealth', 'lifestyle', 'fear', 'hope', 'numbers',
       'gpe', 'person', 'org', 'compound','source_stance_likert_three']].dropna()


x = subset[['pro_environment',
       'extremeweatherevent', 'renewableenergy', 'nonrenewableenergy',
       'wildfire', 'pollution', 'ecosystem', 'sealevelrise', 'politics',
       'finance', 'publichealth', 'lifestyle', 'fear', 'hope', 'numbers',
       'gpe', 'person', 'org', 'compound']]
y = subset['source_stance_likert_three']
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
print(results.summary())
```
